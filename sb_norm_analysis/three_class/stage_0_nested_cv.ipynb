```python
data_path = os.path.join(base_path, files[4])
features = pd.read_csv(data_path)
features.drop(columns=['center_time', 'start_time', 'end_time'], inplace=True)
print(f"Tuning parameters for 2s_no") 

X = features.drop(columns=['label', 'experiment_id'])
y = features['label']
groups = features['experiment_id']
```

```python
# Example usage:
def run_example(X, y, groups):
    """
    Example of how to use the improved NestedCVOptimizer
    """
    # Assuming you have your data loaded
    optimizer = NestedCVOptimizer(
        X=X,
        y=y,
        groups=groups,
        n_outer_folds=5,
        n_inner_folds=3,
        n_trials=50,
        random_state=42
    )
    
    # Run nested CV
    results = optimizer.run_nested_cv()
    
    # Get fold accuracies for analysis
    fold_accuracies = optimizer.get_fold_results()
    print("\nFold accuracies for analysis:")
    for model, accuracies in fold_accuracies.items():
        print(f"{model}: {accuracies}")
    
```

```python
run_example()
```
